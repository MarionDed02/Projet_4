## Processing non-numerical data

On transforme les donn√©es cat√©gorielles des colonnes 'BuildingType' et 'LargestPropertyUseType' en valeurs num√©riques afin de pouvoir les exploiter dans les algorithmes d'apprentissage automatique :

```python
columns_to_encode = ['BuildingType', 'LargestPropertyUseType']

# Cr√©ons un dataframe temporaire contenant ces colonnes
data_to_encode = data[columns_to_encode]

# Instancions l'encodeur OneHotEncoder
encoder = OneHotEncoder()

# On fit et transforme les donn√©es avec l'encodeur
encoded_data = encoder.fit_transform(data_to_encode).toarray()

# Obtenons les noms des colonnes apr√®s l'encodage
encoded_columns = encoder.get_feature_names_out(input_features=columns_to_encode)

# Cr√©ons un nouveau dataframe avec les donn√©es encod√©es et les noms de colonnes
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns, index=data.index)

# Concat√©nons ce dataframe avec le dataframe d'origine
data_num = pd.concat([data, encoded_df], axis=1)

# Supprimons les colonnes d'origine
data_num.drop(columns=columns_to_encode, inplace=True)
```

```python
# S√©lectionnons uniquement les colonnes num√©riques
numeric_columns = data.select_dtypes(include=[np.number])

# It√©rons sur les colonnes num√©riques
for colonne in numeric_columns.columns:
    # Calcul des statistiques
    moyenne = numeric_columns[colonne].mean()
    mediane = numeric_columns[colonne].median()
    ecart_type = numeric_columns[colonne].std()
    variance = numeric_columns[colonne].var()
    minimum = numeric_columns[colonne].min()
    maximum = numeric_columns[colonne].max()
    quantile_25 = numeric_columns[colonne].quantile(0.25)
    quantile_75 = numeric_columns[colonne].quantile(0.75)
    skewness = numeric_columns[colonne].skew()
    kurt = numeric_columns[colonne].kurtosis()
    

    # Affichage des statistiques
    print(f"Statistiques pour '{colonne}':")
    print(f"Moyenne: {moyenne:.2f}")
    print(f"M√©diane: {mediane:.2f}")
    print(f"√âcart Type: {ecart_type:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"Minimum: {minimum:.2f}")
    print(f"Maximum: {maximum:.2f}")
    print(f"Quartile 25%: {quantile_25:.2f}")
    print(f"Quartile 75%: {quantile_75:.2f}")
    print(f"Skewness empirique: {skewness:.2f}")
    print(f"Kurtosis empirique: {kurt:.2f}")   

    # Cr√©ation de l'histogramme
    plt.figure(figsize=(5, 3))
    sns.histplot(data=data, x=colonne, bins=20, kde=False, color='lightblue', edgecolor='k')
    plt.title(f' "{colonne}" - Histogramme', fontsize=12)
    plt.xlabel('Valeurs', fontsize=10)
    plt.ylabel('Fr√©quence', fontsize=10)
    plt.tick_params(axis='both', labelsize=10)
    plt.show()
    

    print("-------------------------------------------------------------------------------------------")
    print(' ')
    print(' ')
```

## Bivariate analysis

G√©n√©rons une heatmap des corr√©lations pour avoir une id√©e des variables corr√©l√©es entre elles :

```python
# S√©lectionnons uniquement les colonnes num√©riques
numeric_data = data.select_dtypes(include=['number'])

# Calculons la matrice de corr√©lation pour les donn√©es num√©riques
corr_matrix = numeric_data.corr()
```

```python
# G√©n√©rons un masque pour le triangle sup√©rieur
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(14, 12))
heatmap = sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', linewidths=0.5, square=True, cbar_kws={"shrink": .5})
plt.title('Heatmap de corr√©lation entre les variables num√©riques', size=18)

# Obtenons les noms des variables
noms_variables = corr_matrix.columns

# Affichons les noms des variables sur l'ace des abscisses √† 45 degr√©s
heatmap.set_xticklabels(noms_variables, rotation=45, ha="right")
plt.show()
```

Les variables les plus corr√©l√©es de notre data set sont :
- Le nombre de b√¢timents, l‚Äô√©lectricit√©, ainsi que la surface plancher. En effet, un nombre de b√¢timents √©lev√© est souvent synonyme d‚Äôune surperficie plus grande. Et cette derni√®re peut entra√Æner plus de besoin en √©lectricit√© si elle est plus √©lev√©e.
