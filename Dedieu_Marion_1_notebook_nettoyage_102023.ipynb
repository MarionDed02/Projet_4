- GHGEmissionsIntensity : Total des émissions de gaz à effet de serre divisé par la surface brute de plancher du bien

Essayons de mieux comprendre l'ENERGY STAR SCORE :

- **Définition :** L'Energy Star Score est un outil développé par l'Agence de protection de l'environnement des États-Unis (EPA) pour évaluer l'efficacité énergétique des bâtiments commerciaux et industriels.
- **Objectif :** Il vise à évaluer et comparer la consommation d'énergie réelle d'un bâtiment par rapport à celle attendue pour un bâtiment similaire en termes de taille et d'utilisation.
- **Échelle de notation :** Le score est attribué sur une échelle de 1 à 100, où un score plus élevé indique une meilleure performance énergétique. La médiane nationale est de 50.
- **Facteurs pris en compte :** Plusieurs variables sont prises en compte, dont la taille du bâtiment, le type d'activité, les heures d'exploitation, etc.
- **Interprétation :** Un score élevé indique une meilleure efficacité énergétique. Un bâtiment avec un score Energy Star élevé est considéré comme plus écoénergétique.
- **Utilité :** Les propriétaires de bâtiments peuvent utiliser ces scores pour identifier des opportunités d'amélioration de l'efficacité énergétique, ce qui peut se traduire par des coûts énergétiques réduits et une empreinte carbone moindre.

## Processing non-numerical data

On transforme les données catégorielles des colonnes 'BuildingType' et 'LargestPropertyUseType' en valeurs numériques afin de pouvoir les exploiter dans les algorithmes d'apprentissage automatique :

```python
columns_to_encode = ['BuildingType', 'LargestPropertyUseType']

# Créons un dataframe temporaire contenant ces colonnes
data_to_encode = data[columns_to_encode]

# Instancions l'encodeur OneHotEncoder
encoder = OneHotEncoder()

# On fit et transforme les données avec l'encodeur
encoded_data = encoder.fit_transform(data_to_encode).toarray()

# Obtenons les noms des colonnes après l'encodage
encoded_columns = encoder.get_feature_names_out(input_features=columns_to_encode)

# Créons un nouveau dataframe avec les données encodées et les noms de colonnes
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns, index=data.index)

# Concaténons ce dataframe avec le dataframe d'origine
data_num = pd.concat([data, encoded_df], axis=1)

# Supprimons les colonnes d'origine
data_num.drop(columns=columns_to_encode, inplace=True)
```

## Univariate analysis

### Quantitative variable

Effectuons l'analyse univariée des variables numériques et calculons pour chaque variable :
- sa moyenne
- sa médiane
- son écart-type
- sa variance
- son minimum
- son maximum
- son premier quartile
- son troisième quartile
- le skewness empirique et le kurtosis empirique pour appréhender les mesures de forme
- et représentons sa distribution avec un histogramme.

```python
# Sélectionnons uniquement les colonnes numériques
numeric_columns = data.select_dtypes(include=[np.number])

# Itérons sur les colonnes numériques
for colonne in numeric_columns.columns:
    # Calcul des statistiques
    moyenne = numeric_columns[colonne].mean()
    mediane = numeric_columns[colonne].median()
    ecart_type = numeric_columns[colonne].std()
    variance = numeric_columns[colonne].var()
    minimum = numeric_columns[colonne].min()
    maximum = numeric_columns[colonne].max()
    quantile_25 = numeric_columns[colonne].quantile(0.25)
    quantile_75 = numeric_columns[colonne].quantile(0.75)
    skewness = numeric_columns[colonne].skew()
    kurt = numeric_columns[colonne].kurtosis()
    

    # Affichage des statistiques
    print(f"Statistiques pour '{colonne}':")
    print(f"Moyenne: {moyenne:.2f}")
    print(f"Médiane: {mediane:.2f}")
    print(f"Écart Type: {ecart_type:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"Minimum: {minimum:.2f}")
    print(f"Maximum: {maximum:.2f}")
    print(f"Quartile 25%: {quantile_25:.2f}")
    print(f"Quartile 75%: {quantile_75:.2f}")
    print(f"Skewness empirique: {skewness:.2f}")
    print(f"Kurtosis empirique: {kurt:.2f}")   

    # Création de l'histogramme
    plt.figure(figsize=(5, 3))
    sns.histplot(data=data, x=colonne, bins=20, kde=False, color='lightblue', edgecolor='k')
    plt.title(f' "{colonne}" - Histogramme', fontsize=12)
    plt.xlabel('Valeurs', fontsize=10)
    plt.ylabel('Fréquence', fontsize=10)
    plt.tick_params(axis='both', labelsize=10)
    plt.show()
    

    print("-------------------------------------------------------------------------------------------")
    print(' ')
    print(' ')
```

## Bivariate analysis

Générons une heatmap des corrélations pour avoir une idée des variables corrélées entre elles :

```python
# Sélectionnons uniquement les colonnes numériques
numeric_data = data.select_dtypes(include=['number'])

# Calculons la matrice de corrélation pour les données numériques
corr_matrix = numeric_data.corr()
```

```python
# Générons un masque pour le triangle supérieur
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(14, 12))
heatmap = sns.heatmap(corr_matrix, mask=mask, cmap='coolwarm', linewidths=0.5, square=True, cbar_kws={"shrink": .5})
plt.title('Heatmap de corrélation entre les variables numériques', size=18)

# Obtenons les noms des variables
noms_variables = corr_matrix.columns

# Affichons les noms des variables sur l'ace des abscisses à 45 degrés
heatmap.set_xticklabels(noms_variables, rotation=45, ha="right")
plt.show()
```

Les variables les plus corrélées de notre data set sont :
- Le nombre de bâtiments, l’électricité, ainsi que la surface plancher. En effet, un nombre de bâtiments élevé est souvent synonyme d’une surperficie plus grande. Et cette dernière peut entraîner plus de besoin en électricité si elle est plus élevée. 
- La consommation d’énergie globale (SiteEUIWN) et les émissions de gaz à effet de serre sont de même corrélés.
